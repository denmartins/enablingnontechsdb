```python
import pandas as pd
```

```python
cars = pd.read_pickle('datasets\\car_original_dataset.pkl')
original_cars = cars.copy(True)
cars
```

```python
type_dummies = pd.get_dummies(cars['type'])
type_dummies
```

```python
df_new = pd.concat([cars, type_dummies], axis=1)
df_new
```

```python
manuf_dummies = pd.get_dummies(cars['manufacturer'])
```

```python
df_new = pd.concat([df_new, manuf_dummies], axis=1)
df_new
```

```python
drive_train_dummies = pd.get_dummies(cars['DriveTrain'])
air_bags_dummies = pd.get_dummies(cars['AirBags'])

df_new = pd.concat([df_new, drive_train_dummies], axis=1)
df_new = pd.concat([df_new, air_bags_dummies], axis=1)
```

```python
df_new.columns
```

```python
columns_removed_after_preprocessing = ['make','manufacturer', 'type', 'DriveTrain', 'AirBags', 'imagepath']
```

```python
preprocessed_columns = [col for col in df_new.columns if not col in columns_removed_after_preprocessing]
preprocessed_columns
```

```python
preprocessed_cars = df_new[preprocessed_columns]
preprocessed_cars
```

```python
preprocessed_cars.to_pickle(path='datasets\\preprocessed_car_dataset.pkl')
```
