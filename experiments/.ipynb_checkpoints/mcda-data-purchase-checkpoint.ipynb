```python
import pandas as pd
from datamanagement.dataaccessobject import DataAccessObject, Dataset
from decision.somselector import SomSelector
from sada.decisionsada import DecisionSADA
# Source: https://scikit-criteria.readthedocs.io/en/latest/index.html
from skcriteria import Data, MIN, MAX
from skcriteria.madm import closeness, simple
```

### Reading data

```python
original_data = pd.read_pickle(os.path.join('datasets', 'uci-original-no-na.pkl'))
original_data
```

```python
preprocessed_data = pd.read_pickle(os.path.join('datasets', 'uci-preprocessed.pkl'))
```

### Creating view

```python
view_columns = [col for col in preprocessed_data.columns
                if not col in ['Name', 'Classification', 'Price', 'Year',
                               'Categorical', 'MissingValues', 'Multivariate']
               ]
working_view = preprocessed_data[view_columns]
```

### Creating the dataset

```python
dataset = Dataset(original_data, working_view)
```

### Configuring the Semiotic Module

```python
# Example selection
selector = SomSelector(som_size=(10,10), learning_rate=1.0, sigma=0.5, num_iterations=1000)

sada = DecisionSADA(dataset,selector=selector)
```

### User input: Data example

```python
example_index = 1
tuple_input = dataset.data_matrix[example_index].tolist()
dataset.original_data.iloc[example_index]
```

### Search/Selection phase

```python
selected_tuples = sada.select(query=tuple_input)
selected_tuples = preprocessed_data.iloc[selected_tuples]
original_data.iloc[selected_tuples.index]
```

### Assessment phase

```python
# Use only the attributes that are important to the decision
columns_assessment_view = ['NumInstances', 'Year', 'MissingValues', 'Price']
assessment_view = selected_tuples[columns_assessment_view]
# Get names of the selected options
names = selected_tuples['Name'].values
```

#### Defining decision maker's criteria and preferences

```python
criteria = [MAX, MAX, MIN, MIN]
weights = [0.3, 0.1, 0.1, 0.5]
```

#### Setting data for MCDA

```python
mcda_data = Data(assessment_view.as_matrix().tolist(), criteria, weights=weights, anames=names, cnames=columns_assessment_view)
```

#### Setting MCDA model

```python
model = closeness.TOPSIS()
decision = model.decide(mcda_data)
decision
```

```python
print("Ideal:", decision.e_.ideal)
print("Anti-Ideal:", decision.e_.anti_ideal)
print("Closeness:", decision.e_.closeness)
```

```python
best_alternative = selected_tuples.iloc[int(decision.best_alternative_)]
original_data.loc[original_data['Name'] == best_alternative['Name']]
```
